# Resumen de la Infografía: Proceso de Predicción de Apertura de Cuenta

Esta infografía ilustra el flujo principal de tu proyecto de modelado predictivo, desde la entrada de datos del cliente hasta la predicción final sobre la apertura de una cuenta bancaria.

## 1. Datos de Entrada (Variables Predictoras)

El modelo utiliza un conjunto de características del cliente, divididas en dos tipos principales:

* **Variables Categóricas:**

  * `Job` (Ocupación)

  * `Marital` (Estado Civil)

  * `Education` (Nivel Educativo)

  * `Housing` (Préstamo Hipotecario)

  * `Loan` (Préstamo Personal)

  * `Contact` (Tipo de Contacto)

  * `Day` (Día del Último Contacto)

  * `Month` (Mes del Último Contacto)

  * `Poutcome` (Resultado de Campaña Anterior)

  * `Hubo_contacto` (¿Fue Contactado Previamente?)

* **Variables Numéricas (ya transformadas logarítmicamente):**

  * `Balance_log` (Balance Promedio Anual Transformado)

  * `Campaign_log` (Contactos en Campaña Actual Transformados)

  * `Pdays_dias_transcurridos_log` (Días Transcurridos Desde Último Contacto Anterior Transformados)

  * `Previous_log` (Contactos Realizados Antes de Esta Campaña Transformados)

## 2. El Modelo de Regresión Logística

Una vez recopilados y preprocesados los datos de entrada, estos pasan por el **Modelo de Regresión Logística**. Este modelo sigue un proceso estadístico para estimar la probabilidad de que un evento binario ocurra (en este caso, que el cliente abra una cuenta).

El corazón de la regresión logística es la función sigmoide, que transforma una combinación lineal de las variables de entrada en una probabilidad entre 0 y 1:

$$
P(Y=1) = \frac{1}{1 + e^{-(\beta_0 + \beta_1 X_1 + \beta_2 X_2 + \dots + \beta_n X_n)}}
$$

Donde:

* $P(Y=1)$ es la probabilidad de que el cliente abra una cuenta.

* $\beta_0$ es el intercepto del modelo.

* $\beta_i$ son los coeficientes para cada variable predictora $X_i$.

* $e$ es la base del logaritmo natural.

Este proceso incluye pasos como la estandarización de variables numéricas y la codificación One-Hot de variables categóricas, asegurando que el modelo reciba los datos en el formato correcto.

## 3. Predicción Final (Abre Cuenta: Sí/No)

El resultado del modelo es una probabilidad. Basándose en un umbral de decisión (por defecto, 0.5), el modelo clasifica al cliente en una de dos categorías:

* **Califica para crear una cuenta:** Si la probabilidad estimada es igual o superior al umbral.

* **No califica para crear una cuenta:** Si la probabilidad estimada es inferior al umbral.

Este flujo simplificado permite a la aplicación tomar decisiones informadas sobre la propensión de un cliente a abrir una cuenta, optimizando así los esfuerzos de marketing y ventas.
