# Situación Problemática: Predicción de Apertura de Cuentas de Ahorro tras Campaña de Marketing

### Contexto del Negocio


Un banco, como parte de su estrategia de crecimiento y captación de clientes,
realiza campañas de marketing directo (principalmente a través de llamadas telefónicas)
para ofrecer a sus clientes y prospectos la apertura de una **cuenta de ahorro a término
(depósito a término)**. Estas campañas implican una inversión considerable de recursos
(tiempo del personal, costes de comunicación, etc.).

## Problema de Negocio


El principal problema al que se enfrenta el banco es la **ineficiencia en la
asignación de recursos** para estas campañas. Actualmente, las llamadas se
realizan a una base de clientes amplia, sin una segmentación efectiva basada
en la propensión real del cliente a aceptar la oferta. Esto lleva a:

- **Altas tasas de no respuesta:** Muchos contactos no resultan en la apertura
  de una cuenta, lo que significa que el tiempo y el esfuerzo de los agentes
  de telemarketing se desperdician.
- **Costos elevados:** Cada llamada y contacto implican un costo, y la baja tasa
  de conversión incrementa el costo por adquisición de cliente.
- **Desgaste del cliente:** Contactar repetidamente a clientes que no están
  interesados puede generar frustración y una percepción negativa del banco.


## Objetivo Analítico (Problema a Resolver)


El banco necesita identificar, de antemano, **qué clientes son más propensos
a suscribirse a un depósito a término** después de ser contactados a través de
una campaña de marketing. Esto implica transformar la ineficiencia actual en
un proceso más dirigido y rentable.


## Características del Problema


Para abordar esta problemática, se dispone de un conjunto de datos histórico de
campañas de marketing, donde cada fila representa un contacto con un cliente.
Las características clave de este problema incluyen:


### 1. Variables de Entrada (Información Disponible):


Se cuenta con información detallada que puede influir en la decisión del cliente, como:

- **Datos demográficos del cliente:** Edad, estado civil, nivel de educación, tipo de trabajo.
- **Situación financiera del cliente:** Saldo promedio anual de su cuenta, si tiene crédito en default,
  si posee un préstamo de vivienda o un préstamo personal.
- **Detalles del Contacto de Marketing:**
    - **Canal y momento:** Tipo de comunicación (teléfono, celular), día y mes del último contacto.
    - **Interacción:** Duración del último contacto (en segundos), número de contactos realizados
      durante esta campaña y en campañas anteriores para el mismo cliente.
    - **Resultado previo:** El resultado de campañas de marketing anteriores para este cliente.


### 2. Variable Objetivo (Lo que se Busca Predecir):


La variable clave es binaria y representa el **resultado de la campaña para cada cliente**:

- **`Sí`**: El cliente ha abierto una cuenta de ahorro a término (depósito a término).
- **`No`**: El cliente no ha abierto una cuenta de ahorro a término.

### 3. Desafíos Inherentes al Problema:


- **Desbalance de Clases:** Es probable que la cantidad de clientes que abren una cuenta (`Sí`)
  sea significativamente menor que la de los que no lo hacen (`No`), lo cual es común en problemas
  de conversión y puede requerir técnicas especiales de manejo de datos.
- **Importancia de la "Duración":** La duración del contacto es una variable altamente predictiva,
  pero solo se conoce *después* de que la llamada ha terminado. Por lo tanto, para un modelo predictivo
  que se use *antes* de contactar al cliente, esta variable debe ser tratada con cautela o excluida
  para evitar el "data leakage" (fuga de información).
- **Multicolinealidad:** Posibles correlaciones entre las variables predictoras que podrían afectar
  la interpretación del modelo.
- **Variables Categóricas:** La necesidad de codificar variables categóricas para que puedan ser
  utilizadas en modelos estadísticos.


La formulación de este problema busca establecer claramente el escenario y
las características de los datos para luego proceder con el análisis y la modelización.
